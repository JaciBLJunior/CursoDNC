```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
credit = pd.read_csv("/content/credit_card_clients.csv")
```

```python
credit
```

```python
credit_group = credit.groupby(["EDUCATION"]).mean().reset_index()
```

```python
# Grafico  matplotlib - Barras
plt.figure(figsize=(10,8))
plt.title("Grafico  matplotlib")
plt.xlabel("Nivel de escolaridade")
plt.ylabel("Média do valor de crédito concedido")
plt.bar(credit_group["EDUCATION"],credit_group["LIMIT_BAL"], color='pink')
```

```python
# Graficos Seaborn - barras
fig,axes = plt.subplots(2,2, figsize=(20,15))
sns.barplot(data=credit_group, x="EDUCATION", y="BILL_AMT1",ax=axes[0,0])
sns.barplot(data=credit_group, x="EDUCATION", y="BILL_AMT2",ax=axes[0,1])
sns.barplot(data=credit_group, x="EDUCATION", y="BILL_AMT3",ax=axes[1,0])
sns.barplot(data=credit_group, x="EDUCATION", y="BILL_AMT4",ax=axes[1,1])
```

```python
# Graficos Plotly - barras
fig2 = go.Figure()
fig2.add_trace(go.Bar(x=credit_group['EDUCATION'],
                      y=credit_group['PAY_AMT1'],
                      marker_color = 'Red',
                      name='Valor pago Janero'))

fig2.add_trace(go.Bar(x=credit_group['EDUCATION'],
                      y=credit_group['PAY_AMT2'],
                      marker_color = 'Gray',
                      name='Valor pago Fevereiro'))

fig2.add_trace(go.Bar(x=credit_group['EDUCATION'],
                      y=credit_group['PAY_AMT3'],
                      marker_color = 'Green',
                      name='Valor pago Fevereiro'))

fig2.add_trace(go.Bar(x=credit_group['EDUCATION'],
                      y=credit_group['PAY_AMT4'],
                      marker_color = 'Blue',
                      name='Valor pago Fevereiro'))

fig2.update_layout(yaxis=dict(title="Valor Pago em 2005",titlefont_size=12,tickfont_size=12),
                   xaxis=dict(title="Grau de escolariedade",titlefont_size=12,tickfont_size=12)
)

fig2.write_html('valorpago.hmtl')
```

```python
# Grafico matpltlib - Linhas
credit_group = credit.groupby(['AGE']).mean().reset_index()
credit_group
```

```python
plt.figure(figsize=(10,8))
plt.title('Grafico de linhas por idade')
plt.xlabel('Idade')
plt.ylabel('Limite')
plt.plot(credit_group['AGE'],credit_group["LIMIT_BAL"])
plt.grid(b=True, color='gray', linestyle='--',linewidth=0.3)
```

```python
# Grafico de linhas Seaborn
fig, axes = plt.subplots(2,2,figsize=(20,10))

sns.lineplot(data=credit, x='AGE',y='PAY_AMT1',hue='SEX',style = 'SEX',palette='Set1',ax=axes[0,0]).set(title='Extrato 1', xlabel='none');
sns.lineplot(data=credit, x='AGE',y='PAY_AMT2',hue='SEX',style = 'SEX',palette='Set2',ax=axes[0,1]).set(title='Extrato 2',xlabel='none');
sns.lineplot(data=credit, x='AGE',y='PAY_AMT3',hue='SEX',style = 'SEX',palette='Set3',ax=axes[1,0]).set(title='Extrato 3',xlabel='none');
sns.lineplot(data=credit, x='AGE',y='PAY_AMT4',hue='SEX',style = 'SEX',palette='Set2',ax=axes[1,1]).set(title='Extrato 4',xlabel='none');
```

```python
# Grafico de linha com a Plotly
fig = go.Figure()
from plotly.subplots import make_subplots

fig = make_subplots(rows=3, cols=2, shared_yaxes=True, subplot_titles=('Abril','Maio','Junho','Julho','Agosto','Setembro'))    

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group["PAY_AMT6"], 
                         line=dict(color='red', width=1.5),name='Valor 1'), row=1,col=1)

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group['PAY_AMT5'], 
                         line=dict(color='blue', width=1.5),name='Valor 2'), row=1,col=2)

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group['PAY_AMT4'], 
                         line=dict(color='green', width=1.5),name='Valor 3'), row=2,col=1)

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group['PAY_AMT3'], 
                         line=dict(color='orange', width=1.5),name='Valor 4'), row=2,col=2)

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group['PAY_AMT2'], 
                         line=dict(color='pink', width=1.5),name='Valor 5'), row=3,col=1)

fig.add_trace(go.Scatter(x=credit_group['AGE'], 
                         y=credit_group['PAY_AMT2'], 
                         line=dict(color='yellow', width=1.5),name='Valor 6'), row=3,col=2)

fig.update_layout(title_text='Valor Pago por meses')
fig.update_yaxes(title_text='Valor dolar 1',row=1,col=1,titlefont_size=8)
fig.update_yaxes(title_text='Valor dolar 2',row=1,col=2,titlefont_size=8)
fig.update_yaxes(title_text='Valor dolar 3',row=2,col=1,titlefont_size=8)
fig.update_yaxes(title_text='Valor dolar 4',row=2,col=2,titlefont_size=8)
fig.update_yaxes(title_text='Valor dolar 5',row=3,col=1,titlefont_size=8)
fig.update_yaxes(title_text='Valor dolar 6',row=3,col=2,titlefont_size=8)

fig.update_xaxes(row=1,col=1,titlefont_size=8)
fig.update_xaxes(row=1,col=2,titlefont_size=8)
fig.update_xaxes(row=2,col=1,titlefont_size=8)
fig.update_xaxes(row=2,col=2,titlefont_size=8)
fig.update_xaxes(title_text='Idade',row=3,col=1,titlefont_size=8)
fig.update_xaxes(title_text='Idade',row=3,col=2,titlefont_size=8)

fig.show()
```

```python
# Grafico com matplotlib - COm 
import numpy as np
t = np.linspace(0.0,2.0,100)
f1 = np.sin(np.pi * t)
f2 = np.sin(2 * np.pi * t)

plt.figure(figsize=(10,10))
plt.plot(t, f1, linewidth=4, label="Função 1")
plt.plot(t, f2, linewidth=8, label="Função 2")
plt.legend()
plt.grid(b=True, color='gray', linestyle='--')
plt.title("Teste")
```
